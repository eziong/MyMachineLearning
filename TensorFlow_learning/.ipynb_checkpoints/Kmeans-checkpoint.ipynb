```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
df = pd.DataFrame(columns=['x','y'])
```

```python
for i in range(30):
    df.loc[i] = [random.randrange(1,20),random.randrange(1,20)]
```

```python
df.head(30)
```

```python
sb.lmplot('x','y',data=df,fit_reg=False,scatter_kws={"s":100})
plt.title('K-means Example')
plt.xlabel('x')
plt.ylabel('y')
```

```python
points = df.values
kmeans = KMeans(n_clusters=4).fit(points) #K-means ++ algorithm
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
df['cluster'] = kmeans.labels_
df.head(30)
```

```python
sb.lmplot('x','y',data=df,fit_reg=False,scatter_kws={"s":150},hue="cluster")
plt.title('K-means Example')
```
